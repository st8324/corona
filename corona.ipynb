```python
%matplotlib inline
import pandas as pd                # 데이터를 저장하고 처리하는 패키지
import matplotlib as mpl           # 그래프를 그리는 패키지
import matplotlib.pyplot as plt    # 그래프를 그리는 패키지
# csv 파일을 읽어서 DataFrame 객체로 만듦. 인덱스 컬럼은 point로 설정
df = pd.read_csv('corona.csv', index_col='day',encoding='utf-8')
df    # df 표시
```

```python
font_name = mpl.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family=font_name)
```

```python
ax = df.plot( title='확진자수', figsize=(12, 4), legend=True, fontsize=12)
ax.legend(['일일확진', '누적확진', '일일사망', '누적사망', '일일해제', '누적해제'], fontsize=12)    # 범례 지정
```
